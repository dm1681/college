```python
Rp = np.linspace(const.R_earth/3, const.R_earth, 1000)
Dp = 2*Rp

rho = 3000 * u.kg/(u.m**3)
Da = 1 * u.km
vimp = 20 * u.km/u.s

Dc = (100 * Da**3 * vimp**2 / (2 * np.pi * const.G * rho * Dp))**(1/4)

fig,ax = plt.subplots(1,1)
ax.set(xlabel='Diameter of Planet', ylabel='Diameter of Crater')
ax.plot(Dp, Dc)
```
