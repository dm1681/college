```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.constants as const
import astropy.units as u

%matplotlib inline
```

```python
# demonstrates the transition probability,
# greatest near w_0, the natural frequency
# plots the Probability Oscillation near w_0,

w = np.linspace(990*np.pi,1010*np.pi,1000)
w_0 = 1000 * np.pi
P = np.absolute(((np.sin((w_0 - w)/2))/(w_0-w)**2))

fig,ax = plt.subplots(1,1)
#fig.set_size_inches(11,11)
ax.plot(w,P)
```

```python
x = np.linspace(0,10,1000)
y = np.linspace(0,10,1000)
phi_gnd = np.exp(-x)
phi_l1_m0 = np.cos(x)
phi_l1_m1 = np.sin(x)*np.exp(1j*y)
fig,ax = plt.subplots(1,1)
ax.plot(x,phi_l1_m1)
ax.plot(-y,phi_l1_m1)
```
