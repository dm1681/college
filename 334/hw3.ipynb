```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u 
import astropy.constants as const
%matplotlib inline
```

```python
C = 0.1e-6 * u.F
f = 2000 * u.Hz
w_c = 2*np.pi * f
R = 1/(w_c*C)
R = R.to(u.Ohm)
w_c.unit
```

```python
fig,ax = plt.subplots(1,1)
fig.set_size_inches(11,8.5)
w = np.linspace(0,100000,100)*u.Hz
Z_out = R/(np.sqrt(1+(R**2 * w**2 * C**2)))
Z_in = np.sqrt(R**2 + 1/(w*C)**2)
ax.plot(w,Z_out)
```
