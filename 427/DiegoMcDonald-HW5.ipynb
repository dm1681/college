```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python

def MC_cir(r, N):
    '''
    This function uses Monte Carlo sampling to estimate pi
    and plots the points generated as well as how the error changes
    after N iterations
    
    Params:
    ------
    r = radius
    N = # of iterations
    
    Output:
    ------
    n_pi = pi estimate after n iterations
    n_err = n_pi - np.pi; the error
    2 plots, described above
    '''
    fig,ax = plt.subplots(1,2)
    fig.set_size_inches(6,3)
    
    ax[0].set(xlim=(0,r), ylim=(0,r), xlabel='x', ylabel='y')
    n = 1
    C = 0
    S = 0
    while n <= N:
        x = random.uniform(0,r)
        y = random.uniform(0,r)
        S+=1
        if (x**2 + y**2) <= (r**2):
            ax[0].scatter(x,y, color='blue', s = 2)
            
            C+=1
        else:
            ax[0].scatter(x,y, color='red', s = 2)
        n_pi = 4. * C/S
        n_err = n_pi - np.pi
        ax[1].scatter(n,n_err, s=2)
        ax[1].set_xscale('log')
       
        n+=1
    return n_pi, n_err

    
```

```python
# takes much longer to run in Python2 than in Python3
MC_cir(1,10000)
```

The graph on the right shows that the error does in fact improve with more iterations (higher N)
